```python
def naive_2mm(p, t, max_mismatches=2):
    occurrences = []
    plen = len(p)
    tlen = len(t)

    for i in range(tlen - plen + 1):
        window = t[i:i+plen]
        mismatches = sum(1 for j in range(plen) if window[j] != p[j])
        if mismatches <= max_mismatches:
            occurrences.append(i)

    return occurrences

# Patrón
pattern = "AGGAGGTT"

# Buscar en la secuencia
matches_approx = naive_2mm(pattern, sequence)

# Obtener offset más a la izquierda
if matches_approx:
    leftmost = min(matches_approx)
    print(f"Offset más a la izquierda con hasta 2 mismatches: {leftmost}")
else:
    print("No se encontraron coincidencias con hasta 2 mismatches.")
```

```python
from collections import defaultdict

def read_reads(filename):
    with open(filename, 'r') as f:
        reads = [line.strip() for line in f if line.strip()]
    return reads

# Función robusta que cuenta bases por posición
def count_bases_by_position(reads):
    position_counts = []
    min_length = min(len(read) for read in reads)  # usamos el más corto

    for i in range(min_length):
        counts = defaultdict(int)
        for read in reads:
            if len(read) > i:  # evitamos IndexError
                counts[read[i]] += 1
        position_counts.append(counts)

    return position_counts

# Leer lecturas desde archivo FASTQ (solo las secuencias)
def extract_sequences_from_fastq(filename):
    sequences = []
    with open(filename, 'r') as f:
        lines = f.readlines()
        for i in range(0, len(lines), 4):
            if i + 1 < len(lines):
                seq = lines[i+1].strip()
                sequences.append(seq)
    return sequences

# Leer secuencias del archivo FASTQ
reads = extract_sequences_from_fastq('ERR037900_1.first1000.fastq')

# Contar bases por posición
base_counts = count_bases_by_position(reads)

# Mostrar frecuencias por posición
for i, counts in enumerate(base_counts):
    total = sum(counts.values())
    print(f"Posición {i}:")
    for base, count in counts.items():
        print(f"  {base}: {count} ({count / total:.2%})")
```
